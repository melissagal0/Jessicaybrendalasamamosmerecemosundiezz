```python
#Registro y Procesamiento de Datos de Ventas
#Queremos administrar una tienda mediante este programa
#Empezamos pidiendo los datos
print ("Buen día, indique cúal es su nombre:")
nombre_vendedor = input()
print ("Ahora indique la clave del producto:")
clave_producto = input()
print("indica la cantidad vendida:")
cantidad_vendida = int(input())
print("Ahora indique el precio del producto:")
precio_producto = float(input())
print("¿Es un cliente frecuente? Escribe nuevo o frecuente en minusculas")
cliente_frecuente = input()

#Calculamos el subtotal que nos ayudara para los dos casos del if
subtotal = cantidad_vendida * precio_producto
print ("Los datos que ingresaste son los siguiente: ")

#Ponemos la condición si se indica que es frecuente el cliente
if cliente_frecuente == "frecuente":
  #Pongo 0.9 por que solo se le va a crobrar el 90 porciento del producto
    descuento = 0.9
    total_con_descuento = subtotal * descuento
    # Se muestran todos los datos que fue agregando el vendedor
    # Y los calculos que se realizaron en el programa
    print ("El nombre del vendedor: " + nombre_vendedor)
    print ("La clave del producto: " + clave_producto)
    #Se convierten en cadenas la cantidad vendidad y el precio del producto
    print ("cantidad vendidad: " + str(cantidad_vendida))
    print ("Precio del producto: " + str(precio_producto))
    print ("¿Es frecuente o nuevo?: " + cliente_frecuente + ", Ya que eres frecuente se te hará un 10% de descuento ")
    print (f"Subtotal: {subtotal}") # Using f-string for printing
    print (f"total con descuento: {total_con_descuento}") # Using f-string for printing

#Si pasa que ponen nuevo, usamos elif, ya que con else no me salía
elif cliente_frecuente == "nuevo":
    total_sin_descuento = subtotal
    # Se muestran todos los datos que fue agregando el vendedor
    # Y los calculos que se realizaron en el programa
    print ("El nombre del vendedor: " + nombre_vendedor)
    print ("La clave del producto: " + clave_producto)
    #Se convierten en cadenas la cantidad vendidad y el precio del producto
    print ("cantidad vendidad: " + str(cantidad_vendida))
    print ("Precio del producto: " + str(precio_producto))
    print ("¿Es frecuente o nuevo?: " + cliente_frecuente + ", ¡Bienvenido! ")
    print (f"total a pagar: {total_sin_descuento}")
```
