```python
import math
#Ecuación Cuadrática (Discriminante Positivo)
#Primero definimos las variables
#Ponemos un if, por si ponen a=0
print ("Este programa calcula el discriminante, mediante una ecuaciòn cuadratica:")
print("Escribe la variable a, nota que debe ser distinta de 0")
mi_variable_a = float(input())

if mi_variable_a == 0:
    print ("¡¡¡Recuerda que debe ser distinta de 0!!!")
else:
    print("Escribe la variable b")
    mi_variable_b = float(input())
    print("Escribe la variable c")
    mi_variable_c = float(input())

    #Escribimos la ecuación completa
    print("La ecuación quedaría de la siguiente forma:")
    print(f"{mi_variable_a}x^2 + {mi_variable_b}x + {mi_variable_c} ")

    #Calculamos el discriminante
    mi_variable_d = mi_variable_b**2 - 4*mi_variable_a*mi_variable_c
    #Se pone f al principio para indicar que es de tipo float
    print (f"El discriminante es: {mi_variable_d}")
    #Ponemos otro if para que solo calcule d > 0
    if mi_variable_d > 0:
        el_valor_x1 = (-mi_variable_b + mi_variable_d**0.5) / (2*mi_variable_a)
        el_valor_x2 = (-mi_variable_b - mi_variable_d**0.5) / (2*mi_variable_a)
        print(f"Las soluciones son reales y distintas: x1 = {el_valor_x1} y x2 = {el_valor_x2}")
    else:
        print("El ejercicio solo resuelve casos con discriminante positivo")
print ("Hasta luego, Gracias por usar mi programa")
```
